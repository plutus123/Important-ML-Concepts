# UCB

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("Ads_CTR_Optimisation.csv")
data.head()
```

```python
data.shape
```

```python
import math
N=10000
d=10
number_of_selection=[0]*d
ads_selected=[]
sum_of_rewards=[0]*d
total_reward=0
for i in range(0,N):
    ad=0
    max_upper_bound=0
    for j in range(0,d):
        if number_of_selection[j] >0:
            average_reward=sum_of_rewards[j] / number_of_selection[j]
            delta_j=math.sqrt(3/2 * math.log(i+1) / number_of_selection[j])
            upper_bound = average_reward + delta_j
        else:
            upper_bound = 1e400
        if(upper_bound > max_upper_bound):
            max_upper_bound= upper_bound
            ad = j
    
    ads_selected.append(ad)
    number_of_selection[ad]+=1
    reward = data.values[i,ad]
    sum_of_rewards[ad] = sum_of_rewards[ad] + reward
    total_reward = total_reward +reward
```

```python
plt.hist(ads_selected)
plt.title("Histogram of ads selection")
plt.xlabel("Ads")
plt.ylabel("Number of times each add was selected")
plt.show()
```

# Thompson Sampling

```python
data=pd.read_csv("Ads_CTR_Optimisation.csv")
data.head()
```

```python
import random 
N=10000
d=10
ads_selected=[]
numbers_of_rewards_1=[0]*d
numbers_of_rewards_0=[0]*d
for n in range(0,N):
    ad=0
    max_random=0
    for i in range(0,d):
        random_beta=random.betavariate(numbers_of_rewards_1[i] + 1 , numbers_of_rewards_0[i] + 1)
        if random_beta > max_random:
            max_random=random_beta
            ad=i
    ads_selected.append(ad)
    reward = data.values[n,ad]
    if reward == 1:
        numbers_of_rewards_1[ad] += 1
    else:
        numbers_of_rewards_0[ad] += 1
    total_reward=total_reward+reward
```

```python
plt.hist(ads_selected)
plt.title("Histogram of ads selection")
plt.xlabel("Ads")
plt.ylabel("Number of times each add was selected")
plt.show()
```
