```python
import pandas as pd
df = pd.read_csv('Data Sheet.csv')
df.head()
```

```python
df['created_at'] = pd.to_datetime(df['created_at'])
df['updated_at'] = pd.to_datetime(df['updated_at'])
df.head()
```

```python
# Calculate delivery time in days
df['delivery_time_days'] = (df['updated_at'] - df['created_at']).dt.days
df.head()
```

```python
# Initialize a DataFrame to hold the final output
final_output = pd.DataFrame()
```

```python
# For average delivery time and TP90 calculation, ensure to group by kind and extract month from 'created_at'
for metric in ['average_delivery_time_days', 'tp90_delivery_time_days']:
    if metric == 'average_delivery_time_days':
        grouped = df.groupby([df['kind'], df['created_at'].dt.month])['delivery_time_days'].mean().unstack(fill_value=0)
    else:  # TP90 delivery time
        grouped = df.groupby([df['kind'], df['created_at'].dt.month])['delivery_time_days'].quantile(0.9).unstack(fill_value=0)
    # Rename columns to 'MMM'YY' format
    grouped.columns = [f'{pd.Timestamp(2023, int(month), 1).strftime("%b")}\'23' for month in grouped.columns]
    # Add a row for metric name
    metric_row = pd.DataFrame([metric], columns=['Product Name'])
    # Concatenate the metric row and the grouped DataFrame
    final_output = pd.concat([final_output, metric_row, grouped.reset_index(drop=True)])
final_output
```

```python
num_customers = df['organization_id'].nunique()
num_customers
```

```python
num_customers_data = pd.DataFrame({'Product Name': ['No. of customers with at least 1 delivery'], '2023': [num_customers]})
num_customers_data
```

```python
final_output = pd.concat([final_output, num_customers_data], ignore_index=True)
final_output = final_output.set_index('Product Name').reset_index()
final_output
```

```python
# Save the final DataFrame to an Excel file
output_file_path = 'final_output_results.xlsx'
final_output.to_excel(output_file_path, index=False)
print(f"Output saved to: {output_file_path}")
```
